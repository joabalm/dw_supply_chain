```python
def garantir_valor_total(df_vendas, df_produtos):
    """
    Garante que valor_total = quantidade * preco_unitario
    usando um merge com a tabela de produtos.
    """

    # 1) Faz um merge para trazer o preco_unitario do produto
    vendas_merged = df_vendas.merge(
        df_produtos[['id_produto', 'preco_unitario']],
        on='id_produto',
        how='left',
        suffixes=('', '_prod')
    )

    # 2) Calcula o valor correto
    valor_correto = vendas_merged['quantidade'] * vendas_merged['preco_unitario']

    # 3) Onde valor_total está diferente ou nulo, substitui
    mask_corrigir = (vendas_merged['valor_total'].isna()) | (vendas_merged['valor_total'] != valor_correto)
    vendas_merged.loc[mask_corrigir, 'valor_total'] = valor_correto[mask_corrigir]

    # 4) Devolve só as colunas originais da df_vendas
    colunas_originais = df_vendas.columns
    df_corrigido = vendas_merged[colunas_originais].copy()

    return df_corrigido
```

```python
df_vendas = garantir_valor_total(df_vendas, df_produtos)
df_vendas.head()
```
